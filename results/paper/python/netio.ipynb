```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

matplotlib.rcParams.update({'font.size': 15})

# bandwidth in MB/s
linux_avg = 3393.5
linux_std = 725.696944705953
hermitux_avg = 65.3
hermitux_std = 4.32177946889679
docker_avg = 355.1
docker_std = 9.19480529671207
osv_avg = 1046
osv_std = 268.629856866284
rump_avg = 95.5815645
rump_std = 9.7274327769141

means = [linux_avg, hermitux_avg, docker_avg, osv_avg, rump_avg]
stdevs = [linux_std, hermitux_std, docker_std, osv_std, rump_std]

fig, ax = plt.subplots()
ax.grid()
ax.set_axisbelow(True)
ax.set_yscale('symlog')#, linthreshy=0.1)

plt.xticks([0, 1, 2, 3, 4,], ["Linux", "HermiTux", "Docker", "OSv", "Rump"], rotation=15)
plt.ylabel("Measured bandwidth\nin MB/s (log. scale)")

rects1 = ax.bar([0, 1, 2, 3, 4], means, yerr=stdevs, color=["skyblue", "moccasin", "lightgray", "lightcoral",\
            "palegreen"], edgecolor=['black', 'black', 'black', 'black', 'black'])

fig = matplotlib.pyplot.gcf()
fig.subplots_adjust(left=0.2)
fig.subplots_adjust(bottom=0.2)
fig.set_size_inches(7,3)

plt.savefig("NetIO.pdf")
```
