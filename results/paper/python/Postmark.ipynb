```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline 

matplotlib.rcParams.update({'font.size': 15})

# in seconds
linux_512_total_avg = 2.6830253
linux_512_tran_avg = 1.874648
linux_512_creation_avg = linux_512_total_avg - linux_512_tran_avg
linux_512_std = 0.004140029791895

hermitux_512_total_avg = 15.693
hermitux_512_tran_avg = 12.253
hermitux_512_creation_avg = hermitux_512_total_avg - hermitux_512_tran_avg
hermitux_512_std = 0.32431294899971

osv_512_total_avg = 51.5469399
osv_512_tran_avg = 44.6178505
osv_512_creation_avg = osv_512_total_avg - osv_512_tran_avg
osv_512_std =2.18482082471932

rump_512_total_avg = 21.4703146
rump_512_tran_avg = 14.9641542
rump_512_creation_avg = rump_512_total_avg - rump_512_tran_avg
rump_512_std = 0.025009303785236

fig, ax = plt.subplots()
ax.grid()
ax.set_axisbelow(True)

plt.ylabel("Postmark execution time (s)")
plt.xticks((0, 1, 2, 3), ("Linux", "HermiTux", "OSv", \
                                 "Rump"))

l1 = mpatches.Patch(facecolor='white', hatch="//", label='File creation\nphase', edgecolor='black')
l2 = mpatches.Patch(facecolor='white', label='Transactions\nphase', hatch="\\\\", edgecolor='black')
plt.legend(handles=[l1, l2], loc="upper left")

rects1 = ax.bar([0], linux_512_creation_avg, hatch="//", color=["skyblue"], \
                edgecolor=['black'])
rects2 = ax.bar([0], linux_512_tran_avg, yerr=linux_512_std, hatch="\\\\", color=["lightblue"], \
                edgecolor=['black'], bottom=linux_512_creation_avg)

rects3 = ax.bar([1], hermitux_512_creation_avg, hatch="//", color=["moccasin"], \
                edgecolor=['black'])
rects4 = ax.bar([1], hermitux_512_tran_avg, yerr=hermitux_512_std, hatch="\\\\", color=["cornsilk"], \
                edgecolor=['black'], bottom=hermitux_512_creation_avg)

rects5 = ax.bar([2], osv_512_creation_avg, hatch="//", color=["lightcoral"], \
                edgecolor=['black'])
rects6 = ax.bar([2], osv_512_tran_avg, yerr=osv_512_std, hatch="\\\\", color=["mistyrose"], \
                edgecolor=['black'], bottom=osv_512_creation_avg)

rects7 = ax.bar([3], rump_512_creation_avg, hatch="//", color=["forestgreen"], \
                edgecolor=['black'])
rects8 = ax.bar([3], rump_512_tran_avg, yerr=rump_512_std, hatch="\\\\", color=["springgreen"], \
                edgecolor=['black'], bottom=rump_512_creation_avg)

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(7,4)

plt.savefig("Postmark.pdf")
```
